```python
colors=['Red','Green', 'Blue','Black','Purple']
states=['WA','NT','SA','Q','V','NSW','T']
neighbours={}
neighbours['WA']=['NT','Q']
neighbours['NT']=['WA','SA','Q']
neighbours['Q']=['NT','SA']
neighbours['SA']=['WA','NT','SA']
neighbours['NSW']=['V','Q','SA']
neighbours['V']=['NSW','SA']
neighbours['T']=[]

colors_of_states={}
def graph_color(state,color):
  for neighbour in neighbours.get(state):
    color_of_neighbour=colors_of_states.get(neighbour)
    if color_of_neighbour==color:
      return False
  return True  

def get_color_of_state(state):
  for color in colors:
      if graph_color(state,color):
        return color
  return None

def main():
  for state in states:
    color=get_color_of_state(state)
    colors_of_states[state]=color
    print(colors_of_states)

main()
```
