```python
#LIST
country=['Bangladesh','India','Nepal','Pakistan']
country1=['USA','UK','Africa','China']
print(country)
print(type(country))
print(country+country1)

#LIST INDEX
print(country[1])

#NEGATIVE LIST INDEX
print(country[-2])

#LIST SILCING
print(country1[1:3])

#Correcting mistake value at List
number=[1,2,3,5]
number[3]=10
print(number)

number[0:3] = [7,8,9]
print(number)

#Appending and Extending
num = [1,2,3]
num.append(4)
print(num)
num.extend([6,9])
print(num)

#Delete
delete=[1,2,3,3]
del delete[3]
print(delete)

```
